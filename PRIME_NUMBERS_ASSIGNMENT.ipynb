```python
from bs4 import BeautifulSoup
import os

# List all files in the input directory
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        file_path = os.path.join(dirname, filename)
        print(f"Processing file: {file_path}")

        # Read HTML content from the file
        with open(file_path, 'r', encoding='utf-8') as file:
            html_content = file.read()

        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(html_content, 'html.parser')

        # Assuming each project detail is within a <tr> tag and fields are continuous
        project_data = []
        current_project = {}
        for row in soup.find_all('tr'):  # Modify this according to the HTML structure
            cells = row.find_all('td')
            if not cells:
                continue

            # Assuming cells[0] contains the field name and cells[1] contains the value
            field_name = cells[0].text.strip()
            field_value = cells[1].text.strip()

            if field_name == 'Name':
                # If there's an ongoing project being collected, save it and start a new one
                if current_project:
                    project_data.append(current_project)
                    current_project = {}
                current_project['Name'] = field_value
            elif field_name == 'GSTIN No.':
                current_project['GSTIN No.'] = field_value
            elif field_name == 'PAN No.':
                current_project['PAN No.'] = field_value
            elif field_name == 'Correspondence Address':
                current_project['Correspondence Address'] = field_value
            elif field_name == 'Permanent Address':
                current_project['Permanent Address'] = field_value

        # Append the last collected project data
        if current_project:
            project_data.append(current_project)

        # Print the extracted projects
        for project in project_data:
            print(f"Name: {project['Name']}")
            print(f"GSTIN No: {project['GSTIN No.']}")
            print(f"PAN No: {project['PAN No.']}")
            print(f"Correspondence Address: {project['Correspondence Address']}")
            print(f"Permanent Address: {project['Permanent Address']}")
            print("-" * 40)
```
